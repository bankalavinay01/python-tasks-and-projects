```python
#Get Dummies
data = {'firstname': ['vinay', 'rohit', 'kohli', 'bumrah', 'shami','surya'], 
        'lastname': ['varma', 'sharma', 'virat', 'jasprit', 'mohammed','yadav'], 
        'employmenttype': ['Student', 'cricket', 'class', 'yorker', 'swing','leg side'],
        'country' :['India','USA','new zealand','england','Australia','south africa']}


df = pd.DataFrame(data, columns = ['firstname','lastname','employmenttype','country'])

print(df)
```

```python
# Apply pd.get_dummies to 'employmenttype' and 'country' columns
df1 = pd.get_dummies(df['employmenttype'], dtype=int)
df2 = pd.get_dummies(df['country'], dtype=int)
print(df1)
print(df2)
```

```python
df
```

```python
frames = [df,df1,df2]
result = pd.concat(frames,axis=1)
print(result)
result.drop(['employmenttype','country'],axis=1,inplace=True)
# inplace=true will perform any action permanently
result
```

```python
#Concatination
import pandas as pd
india_weather = pd.DataFrame({
    "city": ["hyderabad","goa","manali"],
    "temperature": [45,48,3],
    "humidity": [90, 95, 30]
})
india_weather
```

```python
us_weather = pd.DataFrame({
    "city": ["new york","chicago","orlando"],
    "temperature": [21,14,35],
    "humidity": [68, 65, 75]
})
us_weather
```

```python
df = pd.concat([india_weather, us_weather])
df
```

```python
# ignore index
df = pd.concat([india_weather, us_weather], ignore_index=True)
df
```

```python
# concatinate using keys
df = pd.concat([india_weather, us_weather], keys=["india", "us"], ignore_index=True)
df
```
